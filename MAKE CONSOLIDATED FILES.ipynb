```python
#MAKE SURE TO SET THE CORRECT DIRECTORIES FOR YOUR SYSTEM.  
#MAKE CONSOLIDATED FOLDER SHOULD CONTAIN ALL NYSE AND NASDAQ FILES


#CONSOLIDATED MAKE

ConsWorkDir = "/Users/big_u/FINRA DOWNLOAD/MAKE CONSOLIDATED"
ConsDir = "/Users/big_u/FINRA DOWNLOAD/CONSOLIDATED FILES/"

os.chdir(ConsWorkDir)

NQbasename = "FNSQshvol"
NYSEbasename = "FNYXshvol"
CONSOLbasename = "CNMSshvol"
start_date = datetime.date(2010, 1, 1)
end_date = datetime.date(2018, 8, 5)
delta = datetime.timedelta(days=1)

while start_date <= end_date:
    new_date = start_date.strftime("%Y%m%d")
    NQ_current_name = NQbasename+new_date+".txt"
    NYSE_current_name = NYSEbasename+new_date+".txt"
    CONSOL_current_name = CONSOLbasename+new_date+".txt"
    print(NQ_current_name)
    print(NYSE_current_name)
    print(CONSOL_current_name)
    if os.path.exists(NQ_current_name):
        print("exists")
        df1 = pd.read_csv(NQ_current_name, delimiter = "|")
        df2 = pd.read_csv(NYSE_current_name, delimiter = "|")
        df = df1.groupby(['Symbol', 'Date']).sum().add(df2.groupby(['Symbol', 'Date']).sum(), fill_value=0)
        print (df)
        df.to_csv(ConsDir+CONSOL_current_name, sep = "|")
        start_date += delta
    else:
        print("doesn't exist")
        start_date += delta
```
