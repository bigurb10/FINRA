```python
import csv
import datetime
import pandas as pd
import os
import glob
```

```python
# get data file names
path =r'/Users/big_u/FINRA DOWNLOAD/CONSOLIDATED FILES'
filenames = glob.glob(path + "/*.txt")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename, delimiter = "|"))

# Concatenate all data into one DataFrame
big_frame = pd.concat(dfs, ignore_index=True, sort=True)
```

```python
big_frame = big_frame.set_index(['Symbol', 'Date'], drop=True)
```

```python
big_frame=big_frame.sort_values(by=['Symbol', 'Date'])
```

```python
big_frame = big_frame.drop(columns=['Market', 'ShortExemptVolume'])
```

```python
big_frame.head()
```

```python
big_frame.tail()
```

```python
is_NaN = big_frame.isnull()
row_has_NaN = is_NaN.any(axis=1)
rows_with_NaN = big_frame[row_has_NaN]
print(rows_with_NaN)
rows_with_NaN.to_csv('test.txt')
```

```python
big_frame=big_frame.dropna()
```

```python
big_frame.head()
```

```python
big_frame.tail()
```

```python
big_frame['iDIX'] = big_frame['ShortVolume'] / big_frame['TotalVolume']
```

```python
big_frame.head()
```

```python
big_frame['iDIX percentrank'] = big_frame.groupby(level=0)['iDIX'].rank(pct=True)
```

```python
big_frame['5 day average iDIX'] = big_frame.groupby(level=0)['iDIX'].rolling(window=5).mean().values
```

```python
big_frame['5 day average iDIX percentrank'] = big_frame.groupby(level=0)['5 day average iDIX'].rank(pct=True)
```

```python
big_frame['10 day average iDIX'] = big_frame.groupby(level=0)['iDIX'].rolling(window=10).mean().values
```

```python
big_frame['10 day average iDIX percentrank'] = big_frame.groupby(level=0)['10 day average iDIX'].rank(pct=True)
```

```python
big_frame.tail(500)
```
